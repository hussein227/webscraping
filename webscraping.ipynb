```
import requests 
from bs4 import BeautifulSoup
```

```
def get_books(url):
    file_name= 'books of ' + str(url.split('/')[-2]) + '.txt'
    request=requests.get(url)
    soup = BeautifulSoup(request.text,'html.parser')
    Books = soup.find_all('article',attrs={'class':'product_pod'})
    Rating_book={"One":1, "Two":2, "Three":3, "Four":4, "Five":5}
    for book in Books:
        print('Book Name: '+ str(book.find("h3").get_text()) + '\n')
        print('Book Price: ' + str(book.find('p',attrs={'class':'price_color'}).get_text())[2:] + '\n')
        str1= str(book.find('p',attrs={'class':'star-rating'}).get('class')[1])
        print('Book Rating: ' + str(Rating_book[str1]) + '\n')
        print('---------------------------------\n')
        print('---------------------------------\n')
```
